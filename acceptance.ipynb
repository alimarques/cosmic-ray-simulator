# Método da aceitação
É usado para funções mais complexas de se realizar a inversão. O método da aceitação consiste em sortear dois números aleatórios: um para verificar em qual intervalo foi a ocorrência e outro para realizar a inversão com uma parte da função.

Dada $f(x) = \alpha_1\,f_1(x) + \alpha_2\,f_2(x)$ normalizada, me que $\alpha_1+\alpha_2 = 1$, sorteamos $u_1$. Se $u_1<\alpha_1$, será feita a inversão com $f_1(x)$. Caso contrário, com $f_2(x)$.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
N = 100000
np.random.seed(5746)
u_1 = np.random.random_sample(N)
np.random.seed(7801)
u_2 = np.random.random_sample(N)
np.random.seed()
```

### f(x) = A(x+2x^4)
Neste exemplo, $A=\dfrac{5}{74}$, $\alpha_1 = \dfrac{10}{74}$ e $\alpha_2 = \dfrac{64}{74}$.

As funções invertidas ficam:
$$x_1^* = 2(u^*)^{1/2}$$
$$x_2^* = 2(u^*)^{1/5}$$

```python
alfa1 = 10/74
alfa2 = 64/74

xf1 = []
xf2 = []
xf = []
for i in range(len(u_1)):
    # Analisa o intervalo de u_1
    if u_1[i] < alfa1:
        f = 2 * (u_2[i])**(1/2)
        # Adiciona em xf1
        xf1.append(f)
    else:
        f = 2 * (u_2[i])**(1/5)
        # Adiciona em xf2
        xf2.append(f)
    # No final adiciona em xf
    xf.append(f)
```

```python
# Histograma de f1
div = 100
hist_f1 = np.histogram(xf1, bins = div)
y_f1 = hist_f1[0]
y_f1n = y_f1/y_f1[-1]


# Histograma de f2
hist_f2 = np.histogram(xf2, bins = div)
y_f2 = hist_f2[0]
y_f2n = y_f2/y_f2[-1]

# Histograma de f
hist_f = np.histogram(xf, bins = div)
y_f = hist_f[0]
y_fn = y_f/y_f[-1]
```

```python
# Produz x
passo = 0.01
x = np.arange(0, 2+passo, passo)
x = x[0:100]
```

```python
plt.plot(x, y_f1n, ".", color = 'g', label = r'$f_1(x)$')
plt.plot(x, y_f2n, ".", color = 'b', label = r'$f_2(x)$')
plt.plot(x, y_fn, ".", color = 'r', label = r'$f\,(x)$')
plt.legend(loc = 'best', fontsize= 10)
plt.show()
```

```python
# Faz o gráfico de f1
A = 5/74
func1 = A * (x)

# Faz o gráfico de f2
func2 = A * (2*x**4)

# Faz o gráfico da função
func = func1 + func2
```

```python
plt.plot(x, func1, color = 'g', label = r'$f_1(x)$')
plt.plot(x, func2, color = 'b', label = r'$f_2(x)$')
plt.plot(x, func, color = 'r', label = r'$f\,(x)$')
plt.legend(loc = 'best')
plt.show()
```

### f(x) = A(sin(x)+3x^2)
Neste exemplo, $A=\dfrac{1}{2+\pi^3}$, $\alpha_1 = \dfrac{2}{2+\pi^3}$ e $\alpha_2 = \dfrac{\pi^3}{2+\pi^3}$.

As funções invertidas ficam:
$$x_1^* = arccos(1-2u^*)$$
$$x_2^* = \pi (u^*)^{1/3}$$

```python
alfa1 = 2/(2 + (math.pi)**3)
alfa2 = (math.pi)**3/(2 + (math.pi)**3)

# Define as inversas
xf1 = []
xf2 = []
xf = []
for i in range(len(u_1)):
    if u_1[i] < alfa1:
        f = np.arccos(1-2*u_2[i])
        xf1.append(f)
    else:
        f = math.pi * (u_2[i])**(1/3)
        xf2.append(f)
    xf.append(f)
```

```python
# Histograma de f1
div = 100
hist_f1 = np.histogram(xf1, bins = div)
y_f1 = hist_f1[0]

# Histograma de f2
hist_f2 = np.histogram(xf2, bins = div)
y_f2 = hist_f2[0]

# Histograma de f
hist_f = np.histogram(xf, bins = div)
y_f = hist_f[0]
```

```python
plt.plot(y_f1, ".", color = 'g', label = r'$f_1(x)$')
plt.plot(y_f2, ".", color = 'b', label = r'$f_2(x)$')
plt.plot(y_f, ".", color = 'r', label = r'$f\,(x)$')
plt.legend(loc = 'best', fontsize = 10)
plt.show()
```

```python
passo = 0.01
A = 1/(2 + (math.pi)**3)
x = np.arange(0, math.pi+passo, passo)
```

```python
# Faz o gráfico de f1
func1 = A * (np.sin(x))

# Faz o gráfico de f2
func2 = A * (3*x**2)

# Faz o gráfico da função
func = func1 + func2
```

```python
plt.plot(x, func1, color = 'g', label = r'$f_1(x)$')
plt.plot(x, func2, color = 'b', label = r'$f_2(x)$')
plt.plot(x, func, color = 'r', label = r'$f\,(x)$')
plt.legend(loc = 'best', fontsize = 10)
plt.show()
```

### f(x) = A(sin(x)+cos(x))
Neste exemplo, $A=\dfrac{1}{2}$, $\alpha_1 = \dfrac{1}{2}$ e $\alpha_2 = \dfrac{1}{2}$.

As funções invertidas ficam:
$$x_1^* = arcsin(u^*)$$
$$x_2^* = arccos(1-u^*)$$

```python
alfa1 = 1/2
alfa2 = 1/2

# Define as inversas
xf1 = []
xf2 = []
xf = []
for i in range(len(u_1)):
    if u_1[i] < alfa1:
        f = np.arcsin(u_2[i])
        xf1.append(f)
    else:
        f = np.arccos(1 - u_2[i])
        xf2.append(f)
    xf.append(f)
```

```python
# Histograma de f1
div = 100
hist_f1 = np.histogram(xf1, bins = div)
y_f1 = hist_f1[0]
passo = 1.6/len(y_f1)
x1 = np.arange(0, 1.6, passo)

# Histograma de f2
hist_f2 = np.histogram(xf2, bins = div)
y_f2 = hist_f2[0]
passo = 1.6/len(y_f2)
x2 = np.arange(0, 1.6, passo)

# Histograma de f
hist_f = np.histogram(xf, bins = div)
y_f = hist_f[0]
passo = 1.6/len(y_f)
x = np.arange(0, 1.6, passo)
```

```python
plt.plot(x1, y_f1, ".", color = 'g', label = r'$f_1(x)$')
plt.plot(x2, y_f2, ".", color = 'b', label = r'$f_2(x)$')
plt.plot(x, y_f, ".", color = 'r', label = r'$f\,(x)$')
plt.legend(loc = 'best')
plt.show()
```
