```python
import numpy as np
import matplotlib.pyplot as plt
import os
import scipy.stats
path = 'images/'
blue = '#117DF5'
dark_blue = '#086A99'
orange = '#FF7F27'
```

```python
N = 10000
sig = 1
np.random.seed(98)
u1 = np.random.random_sample(N)
np.random.seed(429)
u2 = np.random.random_sample(N)
```

```python
theta = 2*np.pi*u1
r = np.sqrt(-2*sig**2*np.log(1-u2))
```

```python
x = r*np.cos(theta)
y = r*np.sin(theta)

data = []

for i in range(len(x)):
    data.append(x[i])
    data.append(y[i])
    
## Curva da gaussiana
x_test = np.arange(-5,5+0.1,0.1)
f_test = np.exp(-(x_test**2)/(2*sig**2))/(2*np.pi*sig)**0.5

## Valor de f(sig)
f_sig = np.exp(-(sig**2)/(2*sig**2))/(2*np.pi*sig)**0.5

## Mostra sigma
x_sig = [sig, sig]
f_sig_list = [0, f_sig]
x_nsig = [-sig, -sig]
```

```python
bins = 50
plt.hist(data, bins = bins,density = True, edgecolor='white',facecolor=blue, label = 'Números sorteados')    ##histograma normalizado
plt.plot(x_test, f_test, color=orange, label = 'Distribuição gaussiana')    ## curva gaussiana
#plt.plot(x_sig, f_sig_list, '--', color = laranjaShow)    ## grafica sigma
#plt.plot(x_nsig, f_sig_list, '--', color = laranjaShow)    ## grafica sigma
plt.xlabel('x', fontsize = 12)
plt.ylabel('f (x)', fontsize = 14)
plt.xticks(np.arange(-5, 6, 1))
plt.xlim(-5, 5)
plt.legend(loc='best',fontsize=9)
plt.savefig(path+'gaussiana_histograma.png')
```
