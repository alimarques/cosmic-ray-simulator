# Amostragem de números randômicos
Neste arquivo, mostra como gerar números randômicos uniformemente distribuídos entre $[0,1]$ e gerar seu histograma.

```python
# Bibliotecas utilizadas para gerar números randômicos e para plotar histograma
import numpy as np
import matplotlib.pyplot as plt
```

```python
# np.random.random_sample() gera somente 1 número randômico se o parêntesis estiver vazio
n_random = np.random.random_sample()
n_random
```

A cada execução do código, gera-se um novo número randômico. Para criar uma lista de números randômicos e continuar trabalhando com ela, isto é, rodar o código e continuar gerando o mesmo número randômico, precisamos utilizar a função seed do random, que estabelece a semente para gerar números randômicos. Mas lembrar de voltar a seed para vazia.

```python
np.random.seed(453)
print(np.random.random_sample())
np.random.seed()
```

```python
# Para criar N números randômicos, usa-se o for para criar uma lista de listas com diferentes N
N = [10, 100, 1000, 10000, 100000]
n_random = []

for n in N:
    n_random.append(np.random.random_sample(n))
```

### Plotar histograma

```python
# bins é a quantidade de intervalos do histograma
for i in n_random:    
    plt.hist(i, bins = 10, color = 'r')
    plt.xlabel("N", fontsize = 10)
    plt.ylabel("Frequência", fontsize = 10)
    plt.show()
```

Quanto mais números gerados, mais uniforme fica a distribuição, apesar de haver flutuações.

```python
# Calcular a media de cada lista para ver se está como esperado (0,5)
medias = [np.mean(n) for n in n_random]
medias
```

```python
n_random_new = np.random.random_sample(30) *10
n_random_new.round(1)

plt.hist(n_random_new, bins = 10, color = 'r')
plt.xlabel("N", fontsize = 10)
plt.ylabel("Frequência", fontsize = 10)
plt.show()
```
